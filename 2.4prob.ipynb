```python
import numpy as np

age = np.array([23, 23, 27, 27, 39, 41, 47, 49, 50, 52, 54, 54, 56, 57, 58, 58, 60, 61])
fat = np.array([9.5, 26.5, 7.8, 17.8, 31.4, 25.9, 27.4, 27.2, 31.2, 34.6, 42.5, 28.8, 33.4, 30.2, 34.1, 32.9, 41.2, 35.7])

# Calculate the mean
age_mean = np.mean(age)
fat_mean = np.mean(fat)

# Calculate the median
age_median = np.median(age)
fat_median = np.median(fat)

# Calculate the standard deviation
age_std = np.std(age, ddof=1)
fat_std = np.std(fat, ddof=1)

print("Age mean: {:.2f}, median: {:.2f}, standard deviation: {:.2f}".format(age_mean, age_median, age_std))
print("%Fat mean: {:.2f}, median: {:.2f}, standard deviation: {:.2f}".format(fat_mean, fat_median, fat_std))
```

```python
import matplotlib.pyplot as plt

plt.boxplot([age, fat], labels=['Age', '%Fat'])
plt.show()
```

```python
import pandas as pd

# Create DataFrame with age and %fat data
data = pd.DataFrame({
    'age': [23, 23, 27, 27, 39, 41, 47, 49, 50, 52, 54, 54, 56, 57, 58, 58, 60, 61],
    '%fat': [9.5, 26.5, 7.8, 17.8, 31.4, 25.9, 27.4, 27.2, 31.2, 34.6, 42.5, 28.8, 33.4, 30.2, 34.1, 32.9, 41.2, 35.7]
})
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

# Create scatter plot
sns.scatterplot(x='age', y='%fat', data=data)
plt.title('Scatter plot of age and %fat')
plt.xlabel('Age')
plt.ylabel('%fat')
plt.show()

# Create QQ plot
sm.qqplot(data['%fat'], line='s')
plt.title('QQ plot of %fat')
plt.show()
```
